Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

## Examining the Dataset

```python
# Import necssary libraries
import pandas as pd
import numpy as np

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
schools.info()
```

```python
schools.describe()
```

## Schools with Best Math Results
The best math results are at least 80% of the maximum possible score of 800 for [math.]()

```python
# Filter necessary columns for the task
math_schools = schools[['school_name', 'average_math']]
math_schools.head()
```

```python
# Create a column that calculates percentage of math scores
math_schools['average_math_percentage'] = (math_schools['average_math'] / 800) * 100
math_schools.head()
```

```python
len(math_schools['school_name'].unique())
```

```python
# Filter the dataset for best math resutls (at least 80%)
best_math_schools = math_schools[math_schools['average_math_percentage'] >= 80][['school_name', 'average_math']]
best_math_schools.sort_values('average_math', inplace = True, ascending = False)
best_math_schools.head()
```

```python
len(best_math_schools['school_name'].unique())
```

```python
best_math_schools['school_name'].unique()
```

## Top 10 Performing Schools Based on SAT Scores
The best schools are those which have the best total score across the three SAT sections

```python
schools.head()
```

```python
# Filter relevant columns
top_performing = schools[['school_name', 'average_math', 'average_reading', 'average_writing']]
top_performing.head()
```

```python
# Calculate the total SAT score by summing all the scores
top_performing['total_SAT'] = top_performing['average_math'] + top_performing['average_reading'] + top_performing['average_writing']
top_performing.head()
```

```python
# Filter for the top 10 best SAT scores
top_10_schools = top_performing[['school_name', 'total_SAT']].sort_values('total_SAT', ascending = False)
top_10_schools = top_10_schools.iloc[:10]
top_10_schools
```

## NYC Borough with the Largest Standard Deviation in SAT Scores

```python
schools.head()
```

```python
# Create a single SAT score column
borough_analysis = schools.copy()
borough_analysis['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
borough_analysis = borough_analysis[['borough', 'total_SAT']]
borough_analysis.head()
```

```python
# Group by borough and calculate relevant statistics
grouped_borough = borough_analysis.groupby('borough')[['total_SAT']].agg(['count', np.mean, np.std]).round(2)
grouped_borough.columns = ['num_schools', 'average_SAT', 'std_SAT']
grouped_borough
```

```python
# Filter the grouped data to find the borough with the largest std
largest_std_borough = grouped_borough['std_SAT'].idxmax()
largest_std_dev = grouped_borough.loc[[largest_std_borough]]
largest_std_dev
```
